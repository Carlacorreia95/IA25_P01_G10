Parse do dataset de hor√°rios, retornando um dicion√°rio com todas as se√ß√µes.

    Retorna:
        dict: Dicion√°rio com as seguintes chaves:
            - 'head': dict com configura√ß√µes gerais
            - 'cc': dict {turma: [cursos]}
            - 'olw': list de cursos com apenas 1 aula/semana
            - 'dsd': dict {professor: [cursos]}
            - 'tr': dict {professor: [slots indispon√≠veis]}
            - 'rr': dict {curso: sala}
            - 'oc': dict {curso: √≠ndice_aula_online}

```python
def parse_dataset(datasetTxt):
    lines = datasetTxt.strip().split('\n')
    data = {
        'head': {},
        'cc': {},
        'olw': [],
        'dsd': {},
        'tr': {},
        'rr': {},
        'oc': {},
        'rooms': [],
    }

    current_section = None

    for line in lines:
        line = line.strip()

        # Ignorar linhas vazias
        if not line:
            continue

        # Detectar cabe√ßalhos de se√ß√£o
        if line.startswith('#'):
            section_name = line.split()[0][1:]  # Remove o '#'
            current_section = section_name
            continue

        # Ignorar coment√°rios
        if line.startswith('‚Äî') or line.startswith('--'):
            continue

        # Processar dados baseado na se√ß√£o atual
        if current_section == 'cc':
            parts = line.split()
            if parts:
                class_id = parts[0]
                courses = parts[1:]
                data['cc'][class_id] = courses

        elif current_section == 'olw':
            # Esta se√ß√£o parece estar vazia no exemplo
            if line.strip():
                data['olw'].append(line.strip())

        elif current_section == 'dsd':
            parts = line.split()
            if parts:
                teacher = parts[0]
                courses = parts[1:]
                data['dsd'][teacher] = courses

        elif current_section == 'tr':
            parts = line.split()
            if parts:
                teacher = parts[0]
                slots = [int(s) for s in parts[1:]]
                data['tr'][teacher] = slots

        elif current_section == 'rr':
            parts = line.split()
            if len(parts) >= 2:
                course = parts[0]
                room = parts[1]
                data['rr'][course] = room

        elif current_section == 'oc':
            parts = line.split()
            if len(parts) >= 2:
                course = parts[0]
                week_index = int(parts[1])
                data['oc'][course] = week_index

        elif current_section == 'rooms':
            if line.strip():
                data['rooms'].append(line.strip())

    return data
```

L√™ o ficheiro do dataset.

```python
def load_dataset_from_file(datasetTxt):
    try:
        with open(datasetTxt, 'r', encoding='utf-8') as f:
            dataset_text = f.read()
        return parse_dataset(dataset_text)
    except FileNotFoundError:
        print(f"Erro: Ficheiro '{datasetTxt}' n√£o encontrado!")
        return None
    except Exception as e:
        print(f"Erro ao ler ficheiro: {e}")
        return None
```

```python
# ============================
# 1) VARI√ÅVEIS (duas por UC)
# ============================
lesson_vars = []                      # ex: "t01_UC11_L1"
lesson_meta = {}                      # var -> {"class":..., "course":..., "lesson":1|2}
for turma, ucs in dados['cc'].items():
    for uc in ucs:
        for i in (1, 2):              # cada UC tem 2 aulas/semana
            var = f"{turma}_{uc}_L{i}"
            lesson_vars.append(var)
            lesson_meta[var] = {"class": turma, "course": uc, "lesson": i}

print(f"Total de vari√°veis criadas: {len(lesson_vars)}")
```

```python
from constraint import Problem
from collections import Counter, defaultdict

# ===============================================
# CRIAR O PROBLEMA
# ===============================================

problem = Problem()

# ===============================================
# MAPEAR CADA UC AO PROFESSOR
# ===============================================
curso_professor = {}
for prof, cursos in dados['dsd'].items():
    for c in cursos:
        curso_professor[c] = prof


# ===============================================
# ADICIONAR VARI√ÅVEIS (j√° filtrando hor√°rios bloqueados)
# ===============================================
for v in lesson_vars:
    uc = v.split("_")[1]  # extrai o nome da UC (ex: "IA", "IM", etc.)
    slots = set(range(1, 21))  # blocos 1 a 20
    prof = curso_professor.get(uc, None)  # professor da UC, se existir

    # se o professor tiver restri√ß√µes, remove os blocos indispon√≠veis
    if prof and prof in dados['tr']:
        slots -= set(dados['tr'][prof])

    problem.addVariable(v, sorted(list(slots)))

    # debug opcional
    # print(f"{v}: {len(slots)} blocos dispon√≠veis ({prof})")


# Fun√ß√£o auxiliar: mapeia bloco -> dia
def dia_do_bloco(slot):
    return ((slot - 1) // 4) + 1  # blocos 1‚Äì4 = dia 1, etc.

# ===============================================
# RESTRI√á√ÉO: m√°ximo 4 aulas por dia por turma
# ===============================================
def slots_do_dia(d):  # d = 1..5
    start = (d - 1) * 4 + 1
    return list(range(start, start + 4))

# Nova restri√ß√£o: no m√°ximo 3 aulas/dia por turma (1 bloco livre por dia)
for turma in dados['cc']:
    turma_vars = [v for v in lesson_vars if v.startswith(turma + "_")]

    for d in range(1, 6):  # Segunda..Sexta
        free_var = f"__{turma}_D{d}_FREE"
        problem.addVariable(free_var, slots_do_dia(d))

        # Nenhuma aula da turma pode usar o bloco livre desse dia
        for v in turma_vars:
            problem.addConstraint(
                (lambda fv: (lambda s, f: s != fv))(free_var),
                (v, free_var)
            )


# ===============================================
# RESTRI√á√ÉO: disponibilidade dos docentes
# ===============================================
curso_professor = {}
for prof, cursos in dados['dsd'].items():
    for c in cursos:
        curso_professor[c] = prof

for prof, indisponiveis in dados['tr'].items():
    cursos_prof = [c for c, p in curso_professor.items() if p == prof]
    for turma, ucs in dados['cc'].items():
        for uc in ucs:
            if uc in cursos_prof:
                for i in (1, 2):
                    var = f"{turma}_{uc}_L{i}"
                    bloqueados = set(indisponiveis)
                    problem.addConstraint(
                        lambda slot, bloqueados=bloqueados: slot not in bloqueados,
                        (var,)
                    )




# ===============================================
# RESTRI√á√ÉO: aulas online ‚Äî a √∫ltima aula da UC deve ser online
# ===============================================

# Identificar as vari√°veis online
online_vars = []
for curso, li in dados['oc'].items():   # ex: PS -> aula 2
    for turma, ucs in dados['cc'].items():
        if curso in ucs:
            var = f"{turma}_{curso}_L{li}"
            online_vars.append(var)

print("Vari√°veis de aulas online:", online_vars)


# Fun√ß√£o auxiliar para obter o dia de um bloco (1‚Äì20 ‚Üí 1‚Äì5)
def dia_do_bloco(slot):
    return ((slot - 1) // 4) + 1  # blocos 1‚Äì4 = dia 1, etc.


# Aula online (L2) deve acontecer DEPOIS da L1 e em DIA DIFERENTE
def online_after_and_different_day(slot_online, slot_offline):
    dia_online = dia_do_bloco(slot_online)
    dia_offline = dia_do_bloco(slot_offline)
    # online √© depois de offline e n√£o no mesmo dia
    return slot_online > slot_offline and dia_online != dia_offline


for v_online in online_vars:
    turma, uc, _ = v_online.split("_")
    v_offline = f"{turma}_{uc}_L1"
    if v_offline in lesson_vars:
        problem.addConstraint(online_after_and_different_day, (v_online, v_offline))

# ===============================================
# RESTRI√á√ÉO: salas fixas para certas UCs
# ===============================================
salas_fixas = {}
for curso, sala in dados['rr'].items():  # ex: UC14 -> Lab01
    for turma, ucs in dados['cc'].items():
        if curso in ucs:
            for i in (1, 2):
                var = f"{turma}_{curso}_L{i}"
                salas_fixas[var] = sala

print("Salas fixas atribu√≠das:", salas_fixas)


# ===============================================
# RESTRI√á√ÉO: professores e turmas n√£o podem ter sobreposi√ß√£o
# ===============================================
for prof, cursos in dados['dsd'].items():
    prof_vars = [v for v in lesson_vars if any(uc in v for uc in cursos)]
    for i in range(len(prof_vars)):
        for j in range(i + 1, len(prof_vars)):
            problem.addConstraint(lambda a, b: a != b, (prof_vars[i], prof_vars[j]))

for turma in dados['cc']:
    turma_vars = [v for v in lesson_vars if v.startswith(turma + "_")]
    for i in range(len(turma_vars)):
        for j in range(i + 1, len(turma_vars)):
            problem.addConstraint(lambda a, b: a != b, (turma_vars[i], turma_vars[j]))


# ===============================================
# NOVA SEC√á√ÉO: gest√£o de SALAS
# ===============================================
rooms_catalog = list(dados.get('rooms', [])) or ["SalaT", "SalaC", "Lab01"]
room_vars = {}

for var in lesson_vars:
    var_room = f"{var}_ROOM"
    room_vars[var] = var_room

    if var in online_vars:
        problem.addVariable(var_room, ["Online"])
    elif var in salas_fixas:
        problem.addVariable(var_room, [salas_fixas[var]])
    else:
        problem.addVariable(var_room, rooms_catalog)


# Restri√ß√µes: n√£o permitir duas aulas na mesma sala ao mesmo tempo
def no_same_room_same_slot(slot_a, room_a, slot_b, room_b):
    if room_a == "Online" or room_b == "Online":
        return True
    return not (slot_a == slot_b and room_a == room_b)

L = lesson_vars
for i in range(len(L)):
    for j in range(i + 1, len(L)):
        va, vb = L[i], L[j]
        problem.addConstraint(
            no_same_room_same_slot,
            (va, room_vars[va], vb, room_vars[vb])
        )


# ===============================================
# RESTRI√á√ÉO: aulas da mesma UC em dias distintos
# ===============================================

#def dias_diferentes(slot1, slot2):
#    # Cada dia tem 4 blocos ‚Üí blocos 1‚Äì4 = dia 1, 5‚Äì8 = dia 2, etc.
#    return ((slot1 - 1) // 4) != ((slot2 - 1) // 4)

# Para cada turma, garante que L1 e L2 da mesma UC est√£o em dias diferentes
#for turma, ucs in dados['cc'].items():
#    for uc in ucs:
#        var1 = f"{turma}_{uc}_L1"
#        var2 = f"{turma}_{uc}_L2"
#        if var1 in lesson_vars and var2 in lesson_vars:
#            problem.addConstraint(dias_diferentes, (var1, var2))


# ===============================================
# OBTER SOLU√á√ÉO
# ===============================================
solution = next(problem.getSolutionIter(), None)

if not solution:
    print("‚ùå Nenhuma solu√ß√£o vi√°vel encontrada.")
else:
    print("‚úÖ Solu√ß√£o vi√°vel encontrada!\n")
    print("==================================================")
    print("HOR√ÅRIO GERAL (por turma, UC, professor e sala)")
    print("==================================================\n")

    def bloco_para_horario(slot):
        dia_idx = (slot - 1) // 4
        bloco_dia = (slot - 1) % 4 + 1
        dias_semana = ["Segunda", "Ter√ßa", "Quarta", "Quinta", "Sexta"]
        horas = {1: "09h‚Äì11h", 2: "11h‚Äì13h", 3: "14h‚Äì16h", 4: "16h‚Äì18h"}
        return dias_semana[dia_idx], horas[bloco_dia]

    for turma in sorted(dados['cc'].keys()):
        print(f"\nTurma {turma}:")
        print("-" * 60)
        vars_turma = [v for v in sorted(solution.keys()) if v.startswith(turma + "_") and not v.endswith("_ROOM")]

        horario = {}
        for v in vars_turma:
            uc = v.split("_")[1]
            slot = solution[v]
            dia, hora = bloco_para_horario(slot)
            prof = curso_professor.get(uc, "‚Äî")
            sala = solution[room_vars[v]]
            horario.setdefault(dia, []).append((hora, uc, prof, sala))

        for dia in ["Segunda", "Ter√ßa", "Quarta", "Quinta", "Sexta"]:
            if dia in horario:
                print(f"\n  {dia}:")
                for hora, uc, prof, sala in sorted(horario[dia]):
                    print(f"    {hora} ‚Üí {uc} ({sala}) ‚Äî Prof. {prof}")
```

```python
# ===============================================
# EXPORTAR PARA PDF
# ===============================================
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet

if solution:
    print("\nüìÑ A gerar PDF com hor√°rios...")

    pdf_path = "Horario_Geral.pdf"
    doc = SimpleDocTemplate(pdf_path, pagesize=landscape(A4))
    styles = getSampleStyleSheet()
    elements = []

    horas = ["09h‚Äì11h", "11h‚Äì13h", "14h‚Äì16h", "16h‚Äì18h"]
    dias_semana = ["Segunda", "Ter√ßa", "Quarta", "Quinta", "Sexta"]

    for turma in sorted(dados['cc'].keys()):
        elements.append(Paragraph(f"Hor√°rio - Turma {turma}", styles['Heading1']))
        elements.append(Spacer(1, 12))

        horario = defaultdict(list)
        for v in lesson_vars:
            if v.startswith(turma + "_"):
                uc = v.split("_")[1]
                slot = solution[v]
                dia, hora = bloco_para_horario(slot)
                prof = curso_professor.get(uc, "‚Äî")
                sala = solution[room_vars[v]]
                horario[(dia, hora)].append((uc, prof, sala))

        tabela = [["Hora"] + dias_semana]
        for hora in horas:
            linha = [hora]
            for dia in dias_semana:
                celulas = [f"{uc}\n{prof}\n({sala})" for (uc, prof, sala) in horario.get((dia, hora), [])]
                linha.append("\n\n".join(celulas) if celulas else "")
            tabela.append(linha)

        t = Table(tabela, colWidths=[70] + [120]*5, rowHeights=50)
        t.setStyle(TableStyle([
            ('BACKGROUND', (0, 0), (-1, 0), colors.lightgrey),
            ('GRID', (0, 0), (-1, -1), 0.5, colors.grey),
            ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
            ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
            ('FONTSIZE', (0, 0), (-1, -1), 8),
            ('BOTTOMPADDING', (0, 0), (-1, -1), 6),
            ('TOPPADDING', (0, 0), (-1, -1), 6),
        ]))

        elements.append(t)

        if turma != sorted(dados['cc'].keys())[-1]:
            elements.append(PageBreak())

    doc.build(elements)
    print(f"‚úÖ PDF criado com sucesso: {pdf_path}")
```
